```python
import pandas as pd
import numpy as np
import matplotlib as plt
import matplotlib.pylab as plt
import matplotlib.pyplot as plt
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
def make_df(data):
    df = pd.read_csv(data, encoding = 'cp949')
    
    df_dr = df.drop(['계', '전', '답','대'],axis = 1)
    
    df_dr['시군구'] = df['시군구'].str.replace('\(계\)', '')
    
    df_dr.rename(columns = {'년(Annual)':'연도','시도':'행정구역','시군구':'지역명'},inplace = True)


    return df_dr
```

```python
base_df = make_df('data 2/지적통계_국토이용현황(2018~2022).csv')
```

```python
base_df = base_df[base_df['연도'] == 2022]
```

```python
base_df = base_df.iloc[:, 1:]
```

```python
base_df = base_df.groupby('행정구역').mean()
```

```python
def make_gr(df,column):
    
    plt.style.use('seaborn')
    plt.rcParams['font.family'] = 'AppleGothic'
    plt.figure(figsize=(30, 20))
    plt.xticks(rotation = 90, fontsize = 50)
    plt.title(column, fontsize = 15)
    sns.barplot(x = df.index, y = column,data = df)

    plt.legend(loc = 2, bbox_to_anchor = (1,1))
```

```python
for column in base_df.columns:
    make_gr(base_df,column)
```
